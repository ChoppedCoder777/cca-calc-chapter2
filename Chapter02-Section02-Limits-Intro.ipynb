```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define x values, compute f(x)
a = 5
x = np.linspace(-a, a, 400)
y = ((x-1)/(x*x - 1))

# Plot f(x)
plt.plot(x, y, label=r'$\frac{x-1}{x^2 -1}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(-0.5, 2, 400)
y = ((x-1)/(x*x - 1))

# Plot f(x)
plt.plot(x, y, label=r'$\frac{x-1}{x^2 -1}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(0.95, 1.05, 100)
y = ((x-1)/(x*x - 1))

# Plot f(x)
plt.plot(x, y, label=r'$\frac{x-1}{x^2 -1}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(-1.05, 1.05, 100)
y = ((x*x+9)**0.5-3)/(x*x)

# Plot f(x)
plt.plot(x, y, label=r'$\frac{\sqrt{t^2 + 9} - 3}{t^2}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the function f(x)
def f(x):
    return x * np.sqrt(1 + x**-2)

# Choose x values approaching zero from the left
x_left = np.linspace(-1, -0.001, 400)
y_left = f(x_left)

# Plot the function
plt.plot(x_left, y_left, label=r'$f(x)=x\sqrt{1+x^{-2}}$')
plt.title('Limit of $f(x)$ as $x \to 0^-$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

# Check the behavior as x approaches 0 from the left
print("As x approaches 0 from the left, f(x) approaches -1.")
print("So, the limit exists and is -1.")
```

```python
# Define the function f(x)
def f(x):
    return x * np.sqrt(1 + x**-2)

# Choose x values approaching zero from the right
x_right = np.linspace(0.001, 1, 400)
y_right = f(x_right)

# Plot the function
plt.plot(x_right, y_right, label=r'$f(x)=x\sqrt{1+x^{-2}}$')
plt.title('Limit of $f(x)$ as $x \to 0^+$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

# Check the behavior as x approaches 0 from the right
print("As x approaches 0 from the right, f(x) approaches 1.")
print("So, the limit exists and is 1.")
```

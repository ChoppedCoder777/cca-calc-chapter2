```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Problem 21
a = -1
delta = 0.0001
x0 = np.linspace(-3, a, 100)
y0 = x0+3
x1 = np.linspace(a+delta, 3, 100)
y1 = 2**x1

# Plot 
plt.plot(x0, y0, label=r'$x+3$', color='red')
plt.plot(x1, y1, label=r'$2^x$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the function
def f(x):
    return 1/(x+2)

# Values for plotting, avoiding x = -2
x_left = np.linspace(-6, -2 - 0.01, 200)
x_right = np.linspace(-2 + 0.01, 2, 200)

plt.plot(x_left, f(x_left), label=r'$f(x)=\frac{1}{x+2}$', color='blue')
plt.plot(x_right, f(x_right), color='blue')
plt.axvline(x=-2, color='red', linestyle='--', label='x = -2 (discontinuity)')
plt.ylim(-10, 10)
plt.title('Graph of $f(x)=1/(x+2)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()

# Explanation:
print("The function f(x) = 1/(x+2) is discontinuous at x = -2 because the denominator becomes zero, so the function is undefined at the asymptote.") ")
```

```python
# Discontinuity point
a = -1

# 1) Define the left-hand piece for x  -1
x_left = np.linspace(-5, a, 200)
y_left = x_left + 3

# 2) Define the right-hand piece for x > -1
x_right = np.linspace(a + 0.01, 2, 200)
y_right = 2 ** x_right

# Explain the jump discontinuity at x = a = -1:
#    left-hand limit  lim(-1) f(x) = (1) + 3 = 2
#    right-hand limit lim(-1) f(x) = 2^(1)   = 0.5
#   Since 2  0.5, f has a jump discontinuity at x = -1.

# 3) Plot both pieces on the same axes
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x_left,  y_left,  color='red',   label='f(x) = x + 3   for x  1')
ax.plot(x_right, y_right, color='black', label='f(x) = 2^x      for x > 1')

# 4) Mark the vertical line and the two limit-points
ax.axvline(a, color='blue', linestyle='--', label='x = 1 (discontinuity)')
ax.scatter([a], [a + 3],  color='red',   zorder=5)  # left-hand value f(1)=2
ax.scatter([a], [2 ** a], color='black', zorder=5)  # right-hand limit f(1)=0.5

# 5) Finalize the plot
ax.set_title('Piecewise Function with Jump Discontinuity at x = 1')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.grid(True)
ax.legend()

plt.show()

```

```python
#The function is discontinuous at x = -1 because the values from both sides of the graph jump to different numbers instead of meeting.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Define plotting range
x_min, x_max = -2 * np.pi, 2 * np.pi

# 2) Find discontinuity points: sin(x)=1  x = /2 + 2k
k_vals = np.arange(
    np.floor((x_min - np.pi/2) / (2 * np.pi)),
    np.ceil((x_max - np.pi/2) / (2 * np.pi)) + 1
)
disc_points = (np.pi/2) + 2 * np.pi * k_vals
disc_points = disc_points[(disc_points > x_min) & (disc_points < x_max)]

# 3) Build valid x and y arrays
x = np.linspace(x_min, x_max, 10_000)
mask = (1 - np.sin(x)) > 0
x_valid = x[mask]
y_valid = 1 / np.sqrt(1 - np.sin(x_valid))

# 4) Plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x_valid, y_valid, color='blue', label=r'$f(x)=1/\sqrt{1-\sin x}$')

# Mark vertical asymptotes and add blow-up markers
for i, xc in enumerate(disc_points):
    ax.axvline(x=xc, color='red', linestyle='--',
               label='Discontinuity' if i == 0 else None)
    # Add nearby points to show blow-up
    x_near = np.linspace(xc - 0.05, xc - 0.001, 100)
    y_near = 1 / np.sqrt(1 - np.sin(x_near))
    ax.plot(x_near, y_near, color='orange', linewidth=1)

# 5) Finalize the graph
ax.set_xlim(x_min, x_max)
ax.set_title('Discontinuities of $f(x)=1/\\sqrt{1-\\sin x}$')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.grid(True)
ax.legend()

# 6) Show the plot
plt.show()

# 7) Print discontinuity locations
print("Vertical asymptotes at x = /2 + 2k within range:")
for xc in disc_points:
    print(f"  x = {xc:.3f}")


```

```python
import numpy as np
import matplotlib.pyplot as plt

def piecewise_f(x):
    """Piecewise definition of f(x)."""
    if x < 0:
        return x + 2
    elif x <= 1:
        return np.exp(x)
    else:
        return 2 - x

def one_sided_limits_and_continuity():
    """Compute one-sided limits and check continuity at x=0 and x=1."""
    # x = 0
    left_0  = (0 - 1e-6) + 2        # approx f(0-)
    right_0 = np.exp(0 + 1e-6)      # approx f(0+)
    f0      = piecewise_f(0)
    print("At x = 0:")
    print(f"  left limit  {left_0:.6f}, right limit  {right_0:.6f}, f(0) = {f0:.6f}")
    print("  Continuous from the right only\n" if left_0 != f0 and right_0 == f0 else "")

    # x = 1
    left_1  = np.exp(1 - 1e-6)      # approx f(1-)
    right_1 = 2 - (1 + 1e-6)        # approx f(1+)
    f1      = piecewise_f(1)
    print("At x = 1:")
    print(f"  left limit  {left_1:.6f}, right limit  {right_1:.6f}, f(1) = {f1:.6f}")
    print("  Continuous from the left only\n" if left_1 == f1 and right_1 != f1 else "")

def plot_piecewise():
    """Sketch the piecewise function with open/closed circles at boundaries."""
    # build x ranges
    x1 = np.linspace(-5, 0, 400, endpoint=False)      # x < 0
    x2 = np.linspace(0, 1, 400)                       # 0  x  1
    x3 = np.linspace(1, 5, 400, endpoint=False)       # x > 1

    y1 = x1 + 2
    y2 = np.exp(x2)
    y3 = 2 - x3

    fig, ax = plt.subplots(figsize=(8, 5))

    # plot each piece
    ax.plot(x1, y1, color='red',   label='x + 2, x < 0')
    ax.plot(x2, y2, color='green', label='e^x, 0  x  1')
    ax.plot(x3, y3, color='blue',  label='2  x, x > 1')

    # circles at x=0
    ax.scatter([0], [2], facecolors='none', edgecolors='red',   s=100, label='(0,2) open')
    ax.scatter([0], [1], facecolors='green', edgecolors='green', s=100, label='(0,1) closed')

    # circles at x=1
    ax.scatter([1], [np.exp(1)], facecolors='green', edgecolors='green', s=100, label='(1,e) closed')
    ax.scatter([1], [1], facecolors='none',  edgecolors='blue',  s=100, label='(1,1) open')

    ax.set_title('Piecewise Function with Discontinuities at x=0 and x=1')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.grid(True)
    ax.legend(loc='upper right')

    plt.show()

if __name__ == '__main__':
    one_sided_limits_and_continuity()
    plot_piecewise()

```

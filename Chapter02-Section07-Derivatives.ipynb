```python
import numpy as np
import matplotlib.pyplot as plt

# Define curve and tangent
def curve(x): 
    return x**2 + 3*x

def tangent(x):
    return x - 1  # y = x - 1 is tangent at x0 = -1

# Prepare plot range (centered at (-1, -2))
x = np.linspace(-5, 3, 400)
y_curve = curve(x)
y_tan   = tangent(x)

# Plot
plt.figure(figsize=(6, 6))
plt.plot(x, y_curve, label='y = x² + 3x', color='blue')
plt.plot(x, y_tan,   label='Tangent at (-1,-2)', color='red', linestyle='--')
plt.scatter([-1], [-2], color='black')
plt.title('Broad View Around (–1, –2)')
plt.xlim(-5, 3)
plt.ylim(-10, 10)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()

```

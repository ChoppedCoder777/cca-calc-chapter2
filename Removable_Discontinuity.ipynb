```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 1 / (x - 2)

# Create x values, avoiding x = 2 to prevent division by zero
x_left = np.linspace(1.5, 1.99, 500)
x_right = np.linspace(2.01, 2.5, 500)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_left, f(x_left), 'b', label='x → 2⁻')
plt.plot(x_right, f(x_right), 'r', label='x → 2⁺')

# Add vertical asymptote at x = 2
plt.axvline(x=2, color='k', linestyle='--', label='x = 2 (asymptote)')

# Add labels and legend
plt.title('Graph of f(x) = 1 / (x - 2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-20, 20)  # Limit y-axis for better visualization
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
